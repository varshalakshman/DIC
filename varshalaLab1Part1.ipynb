```r
# Function for printing out all the squares upto x
generateSquares <- function(x) {
  for(i in 1:x) {
    y <- i^2
    print(y)
  }
  return(x^2)
}
a = generateSquares(4)
print(a)
# Creating a function that takes arguments
func_arguments <- function(a=1,b=2,c=3) {
  res <- a + (b * c)
  print(res)
}
#calling by default
func_arguments()
# Calling by position
func_arguments(4,5,6)
# Calling by names
func_arguments(a = 4, b = 5, c = 3)
```

## Boxplots :
For a distribution of numerical data, there are 5 numbers of significance. The lower bound - L, upper bound - U, first quarter 
upper bound - Q1, third quarter lower bound - Q3 and the median/center of the distribution - M.
These 5 numbers are represented using Box plots.

In the above box plot, we can compare the box plots of 2 different distributions. Both the distributions have the same range but 
the first data set has lower values. First distribution has a median of 34, which is lower than the median of the second 
distribution (whcih is at 45). Although both data sets have same range, the first data set has a more wider distribution than 
the second one, for the Q1 to Q3 region
